```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
np.random.seed(0)
centroids = [[3,3], [-3, -2], [2, -3], [0, 0]]
X, y = make_blobs(n_samples=5000, centers=centroids, cluster_std=0.8)
```

```python
plt.scatter(X[:, 0], X[:, 1], marker='.')
plt.show()
```

```python
k_means = KMeans(init = "random", n_clusters = 4, n_init = 15)
```

```python
k_means.fit(X)
```

```python
k_means.labels_
```

```python
k_means.cluster_centers_
```

```python
print(centroids)
```

```python
print(np.round(k_means.cluster_centers_))
```

```python
fig = plt.figure(figsize=(8, 8))
# Har bir klaster uchun alohida rang
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means.labels_))))
ax = fig.add_subplot(1, 1, 1)

for k, col in zip(range(len([[3,3], [-3, -2], [2, -3], [0, 0]])), colors):

    # Create a list of all data points, where the data points that are
    # in the cluster (ex. cluster 0) are labeled as true, else they are
    # labeled as false.
    my_members = (k_means.labels_ == k)

    # Define the centroid, or cluster center.
    cluster_center = k_means.cluster_centers_[k]

    # Plots the datapoints with color col.
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')

    # Plots the centroids with specified color, but with a darker outline
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)

# Title of the plot
ax.set_title('KMeans')

# Remove x-axis ticks
ax.set_xticks(())

# Remove y-axis ticks
ax.set_yticks(())

# Show the plot
plt.show()
```

```python
# KMeans algoritmini turli k qiymatlarida sinab ko'ramiz
k_values = [2, 3, 4, 5, 6]

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X)

    # Klaster markazlari
    centers = kmeans.cluster_centers_

    # Har bir nuqtaning klasterlarga tegishliligini chiqaramiz
    labels = kmeans.labels_

    # Klasterlar va markazlarni chizish
    plt.figure(figsize=(6, 4))
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')
    plt.title(f"K={k} bo'lgan KMeans")
    plt.show()
```

```python
# K qiymatlari bo'yicha KMeans algoritmini ishga tushirish va vizualizatsiya qilish
k_values = [2, 3, 4, 5]  # turli klasterlar soni

for k in k_values:
    # KMeans algoritmini ishga tushiramiz
    k_means = KMeans(n_clusters=k, init='k-means++', random_state=42)
    k_means.fit(X)

    # Grafik yaratish
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)

    # Har bir klaster uchun alohida rang tayinlash
    colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means.labels_))))

    for k_num, col in zip(range(k), colors):
        # Klaster a'zolarini aniqlash
        my_members = (k_means.labels_ == k_num)

        # Klaster markazini aniqlash
        cluster_center = k_means.cluster_centers_[k_num]

        # Klasterdagi nuqtalarni chizish
        ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')

        # Klaster markazini chizish
        ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=10)

    # Grafik nomini kiritish
    ax.set_title(f'KMeans with k={k}')

    # O'qlarni o'chirish
    ax.set_xticks(())
    ax.set_yticks(())

    # Grafikni ko'rsatish
    plt.show()
```
