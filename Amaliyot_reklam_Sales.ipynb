```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/reklama_sales.csv')

x_tv = df['TV']
y = df['Sales']
x_tv = sm.add_constant(x_tv)

model_tv = sm.OLS(y,x_tv).fit()
print("Tv uchun model natijalari")
print(model_tv.summary())

plt.figure(figsize=(10,5))
plt.scatter(df['TV'], df['Sales'], color='blue', label ='Malumotlar')
plt.plot(df['TV'], model_tv.predict(x_tv), color='red', label='Regression Line')
plt.xlabel("TV reklamasi")
plt.ylabel("Daromad (Sales)")
plt.title("Oddiy chiziqli regressiya modeli (TV)")
plt.legend()
plt.show()


x_radio = df['Radio']
x_radio = sm.add_constant(x_radio)

model_radio = sm.OLS(y,x_radio).fit()
print("Radio uchun model natijalari")
print(model_radio.summary())

plt.figure(figsize=(10,5))
plt.scatter(df['Radio'], df['Sales'], color='orange', label = 'Malumotlar')
plt.plot(df['Radio'], model_radio.predict(x_radio), color = 'red', label = 'Regression Line')
plt.xlabel("Radio reklamasi ")
plt.ylabel("Daromad (sales)")
plt.legend()
plt.show()



x_newspaper = df['Newspaper']
x_newspaper = sm.add_constant(x_newspaper)
model_newspaper = sm.OLS(y,x_newspaper).fit()
print("Newspaper uchun model natijalari ")
print(model_newspaper.summary())


plt.figure(figsize=(10,5))
plt.scatter(df['Newspaper'], df['Sales'], color= 'green', label = 'Malumotlar')
plt.plot(df['Newspaper'], model_newspaper.predict(x_newspaper), color = 'red', label = 'Regression Line')
plt.xlabel("Newspaper reklamasi ")
plt.ylabel("Daromad (sales)")
plt.legend()
plt.show()



```

```python
print("TV R-kvadrat:", model_tv.rsquared)
print("Radio R-kvadrat:", model_radio.rsquared)
print("Newspaper R-kvadrat:", model_newspaper.rsquared)
```

```python
results = pd.DataFrame({
    'Model': ['TV', 'Radio', 'Newspaper'],
    'R-squared': [model_tv.rsquared, model_radio.rsquared, model_newspaper.rsquared],
    'p-value': [model_tv.pvalues[1], model_radio.pvalues[1], model_newspaper.pvalues[1]]
})

print(results)
```
